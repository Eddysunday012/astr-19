```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(0, 1, 101)
```

```python
def sinFunc(x):
    return math.sin(x)
```

```python
def cosFunc(x):
    return math.cos(x)
```

```python
sinF = np.vectorize(sinFunc)
cosF = np.vectorize(cosFunc)
```

```python
ySin = sinF(x)
yCos = cosF(x)
```

```python
fig = plt.figure()


fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Sin(x) vs Cos(x)')
ax1.plot(x, ySin)
ax1.set_xlabel("x")
ax1.set_ylabel("sin(x)")
ax2.plot(x, yCos)
ax2.set_xlabel("x")
ax2.set_ylabel("cos(x)")
plt.show()
```
