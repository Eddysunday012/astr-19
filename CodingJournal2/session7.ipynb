```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 101)
```

```python
y = np.exp(x)
```

```python
fig = plt.figure()


plt.plot(x, y, 'r')
plt.xlim(0, 1)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.show()
```

```python
plt.savefig("Exp_plot.pdf")
```
