```python
from math import sin, cos, pi
from tabulate import tabulate
```

```python
def sinFunc(x):
    return sin(x)
```

```python
def cosFunc(x):
    return cos(x)
```

```python
def getTable(func):
    topNum = 1000
    mainList = []
    for i in range(topNum+1):
        val = (i/topNum) * 2*pi
        mainList.append([i, val, func(val)])
    return (tabulate(mainList, headers=["index", "value", "function(x)"]))
```

```python
print(getTable(sinFunc))
```

```python
print(getTable(cosFunc))
```
